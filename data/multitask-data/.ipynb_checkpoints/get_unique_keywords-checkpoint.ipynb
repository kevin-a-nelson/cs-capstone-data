```python
df.head()
```

```python
unique_keywords = df['keywords'].unique()
```

```python
print(unique_keywords)
```

```python
for i in range(len(unique_keywords)):
    keystring = unique_keywords[i]
    keystring = keystring.translate(str.maketrans('', '', '[]\''))
    unique_keywords[i] = keystring
```

```python
final_arr = []
```

```python
for i in range(len(unique_keywords)):
    temp_list = unique_keywords[i].split(', ')
    for keyword in temp_list:
        if keyword not in final_arr:
            final_arr.append(keyword)
```

```python
final_arr.sort()
```

```python
print(final_arr)
```
